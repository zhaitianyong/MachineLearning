```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1, 3, 2, 3, 5], dtype=float)
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

x_mean, y_mean
```

# 最小二乘法推导的公式直接计算




```python
 a = np.sum((x-x_mean)*(y-y_mean))/(np.sum((x-x_mean)**2))
```

```python
 a
```

```python
b = y_mean - a * x_mean
```

```python
a, b
```

```python
x1 = [0, 6]
y1 = [b, a * 6 + b]
```

```python
plt.scatter(x, y)
plt.plot(x1, y1)
plt.axis([0, 6, 0, 6])
plt.show()
```
