```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python


def logistic(X, y, W, b):
    m, n = X.shape
    e = sigmoid(X.dot(W) + b)
    cost = -np.sum(y*np.log(e) + (1-y)*np.log(1-e))/m   
    dy =  e-y
    dW =  (X.T).dot(dy) / m
    db = np.sum(dy) / m
    return cost, dW, db
    
    
```

```python
def logistic_train(X_train, y_train, lr, epochs):
    m, n = X_train.shape
    # 初始化参数
    W = np.zeros((n,1))
    b = 0
    for i in range(epochs):
        cost, dW, db = logistic(X_train, y_train, W, b)
        
        W = W - lr*dW
        b = b - lr*db
        
        print("epoch:", i, " cost:", cost)
        
    return W, b
    
```

```python
def predict(X_test, W, b):
    y_predict = sigmoid(X_test.dot(W) + b)
    for i in range(len(y_predict)):
        if y_predict[i] > 0.5:
            y_predict[i] = 1
        else:
            y_predict[i] = 0
    return y_predict
```

```python
# 准备数据
from sklearn.datasets.samples_generator import make_classification
X,labels=make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=2)
rng=np.random.RandomState(2)
X+=2*rng.uniform(size=X.shape)

unique_lables=set(labels)
colors=plt.cm.Spectral(np.linspace(0, 1, len(unique_lables)))
for k, col in zip(unique_lables, colors):
    x_k=X[labels==k]
    plt.plot(x_k[:, 0], x_k[:, 1], 'o', markerfacecolor=col, markeredgecolor="k",
             markersize=14)
plt.title('data by make_classification()')
plt.show()
```
